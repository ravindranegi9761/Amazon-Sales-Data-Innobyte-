# Visualization

```python
total_cancelled_per_state = df.groupby('ship-state')['Status'].apply(lambda x: (x == 'Cancelled').sum())
a=total_cancelled_per_state.sort_values(ascending=False)
a
```

```python
a.plot(kind='bar',title='no of cancelled order by ship state',xlabel='state',ylabel='no. of order',figsize=(10,6))
```

```python
average_cancelled_per_state = df.groupby('ship-state')['Status'].apply(lambda x: (x == 'Cancelled').mean())
a=average_cancelled_per_state.sort_values(ascending=False)
```

```python
a.plot(kind='bar',title='average of cancelled order by ship state',xlabel='state',ylabel='average',figsize=(10,6),color='green')
```

```python
b = df.groupby('Fulfilment')['Status'].apply(lambda x: (x == 'Cancelled').mean())
b
```

# Months wise revenue

```python
revenue_m_w=df.groupby('Month')['Amount'].sum()
revenue_m_w
```

```python
ax=revenue_m_w.plot(kind='line',figsize=(10, 6))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.xlabel('month')
plt.ylabel('revenue')
plt.title('Revenue by Months')
plt.show()
```

```python
df_april = df[df['Month'] == 'April']
total_amount_april = df_april['Amount'].sum()
print("total revenue in april month =",total_amount_april)
df_may = df[df['Month'] == 'May']
total_amount_may = df_may['Amount'].sum()
print('total revenue in may month =',total_amount_may)
df_june = df[df['Month'] == 'June']
total_amount_june = df_june['Amount'].sum()
print('total revenue in june month =',total_amount_june)
april_to_may=(total_amount_april-total_amount_may)/total_amount_april*100
print("percent of revenue decreased from april to may =",april_to_may)
may_to_june=(total_amount_may-total_amount_june)/total_amount_may*100
print("percent of revenue decreased from may to june =",may_to_june)
```
